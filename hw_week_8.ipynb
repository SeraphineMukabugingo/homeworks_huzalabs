```python
# Load data from Excel file
df = pd.read_excel('Rwanda_Maize_Production.xlsx', engine='openpyxl')
df
```

```python
# Explore data
print(df.info())
```

```python
print(df.tail())
```

```python
# checking the missing values
df.isna()
```

```python
# removing the total row bcs is not necessary in data analysis
exclude_total = df[df['District'] != 'Total']
exclude_total
```

```python
# calculating descriptive statistics
exclude_total.describe()
```

```python
# Identify top maize producing districts
sorted_df =exclude_total.sort_values(by='Total Production (Ton)', ascending=False)
sorted_df
```

```python
# Bar plot to visualize maize production by district
plt.figure(figsize=(15, 8))
plt.bar(sorted_df['District'], sorted_df['Total Production (Ton)'], color='green', edgecolor='black')
plt.title('Maize Production by District')
plt.xlabel('District')
plt.ylabel('Total Production (Ton)')
plt.xticks(rotation=90)
plt.show()

```

```python
# Visualize distribution of maize production
plt.figure(figsize=(10, 6))
plt.hist(sorted_df['Total Production (Ton)'], bins=20, color='violet', edgecolor='black')
plt.title('Distribution of Maize Production')
plt.xlabel('Production Value (Ton)')
plt.ylabel('Frequency')
plt.show()
```

```python
# showing a boxplot of a dataset
plt.figure(figsize=(10, 6))
plt.boxplot(sorted_df['Total Production (Ton)'])
plt.title('Boxplot of Maize Production')
plt.ylabel('Production Value (Ton)')
plt.show()
```

```python
# Load the shapefile
gdf = gpd.read_file('District.shp')
gdf.info()
```

```python
gdf
```

```python
# Merge maize production data with geospatial data
merged = gdf.set_index('District').join(sorted_df.set_index('District'))

# Plot the map
plt.figure(figsize=(12, 8))
merged.plot(column='Total Production (Ton)', cmap='OrRd', legend=True)
plt.title('Maize Production by District in Rwanda (2018)')
plt.show()
```

```python
# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged.plot(column='Total Production (Ton)', cmap='OrRd', legend=True, ax=ax)

# Add district names
for x, y, label in zip(merged.geometry.centroid.x, merged.geometry.centroid.y, merged.index):
    ax.annotate(text=label, xy=(x, y), xytext=(3, 3),
                textcoords="offset points", fontsize=8, color='black')

ax.set_title('Maize Production by District in Rwanda (2018)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Correct the district names in the dataframe to match the shapefile
sorted_df['District'] = sorted_df['District'].replace({'Gigasara': 'Gisagara', 'Nyagatere': 'Nyagatare'})
```

```python
# Check for matching district names
print("Shapefile Districts:", gdf['District'].unique())
print("Data Districts:", sorted_df['District'].unique())

# Merge maize production data with geospatial data
merged = gdf.set_index('District').join(sorted_df.set_index('District'))

# Plot the map
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged.plot(column='Total Production (Ton)', cmap='OrRd', legend=True, ax=ax,
            legend_kwds={'label': "Total Production (Ton)",
                         'orientation': "horizontal"})

# Add district names
for x, y, label in zip(merged.geometry.centroid.x, merged.geometry.centroid.y, merged.index):
    ax.annotate(text=label, xy=(x, y), xytext=(3, 3),
                textcoords="offset points", fontsize=8, color='black')

# Manually extend plot limits to ensure all districts are included
x_buffer = 0.1
y_buffer = 0.1
ax.set_xlim(merged.total_bounds[0] - x_buffer, merged.total_bounds[2] + x_buffer)
ax.set_ylim(merged.total_bounds[1] - y_buffer, merged.total_bounds[3] + y_buffer)

# Customize plot
ax.set_title('Maize Production by District in Rwanda (2018)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
